# Exploratory Data Analysis

```python
ax=sns.countplot(x='Size',data=df)
```

**From above Graph you can see that most of the people buys M-size**

# Group By
**The groupby() function in pandas is used to group data based on one or more columns in a 
DataFrame**

```python
df.groupby(['Size'],as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)
```

```python
S_Qty=df.groupby(['Size'],as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False)

sns.barplot(x='Size',y='Quantity',data=S_Qty)
```

**From above Graph you can see that most of the Qty buys M-Size in the sales**

```python
#Histogram
df['Size'].hist()
```

```python
# As I have to draw graph of 'Category' I have to change the data type
df['Category']=df['Category'].astype(str)
column_data=df['Category']
plt.figure(figsize=(10,5))
plt.hist(column_data, bins=30, edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
```

**From above Graph you can see that most of the buyerrs are T-shirt**

```python
# Checking the B2B by using pie chart
B2B_Check=df['B2B'].value_counts()

# plot the pie chart

plt.pie(B2B_Check,labels=B2B_Check, autopct='%1.1f%%')

#plt.axis('equal')
plt.show()
```

```python
# Checking the B2B by using pie chart
B2B_Check=df['B2B'].value_counts()

# plot the pie chart

plt.pie(B2B_Check,labels=B2B_Check.index, autopct='%1.1f%%')

#plt.axis('equal')
plt.show()
```

**From above graph we can see that maximum i.e 99.3% of buyers are retailers and 0.8% are B2B buyers**

```python
# prepare data for scatter plot
x_data = df['Category']
y_data = df['Size']

# plot the scatter plot
plt.scatter(x_data,y_data)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

```python
# plot count of cities by state
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
# top_10_States
top_10_States = df['ship-state'].value_counts().head(10)
# plot count of cities by state
plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top_10_States.index)], x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=45)
plt.show()
```

**From above Graph you can see that most of the buyers are Maharashtra state**
